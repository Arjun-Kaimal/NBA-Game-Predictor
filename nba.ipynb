```python
games["GAME_DATE_EST"] = pd.to_datetime(games["GAME_DATE_EST"])
```

```python
games["OPP_CODE"] = games["VISITOR_TEAM_ID"].astype("category").cat.codes
```

```python
games["DAY_CODE"] = games["GAME_DATE_EST"].dt.dayofweek
```

```python
games
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=50, min_samples_split=10, random_state=1)
train = games[games["GAME_DATE_EST"] < '2018-01-01']
test = games[games["GAME_DATE_EST"] >= '2018-01-01']
predictors = ["OPP_CODE", "DAY_CODE"]
```

```python
rf.fit(train[predictors], train["HOME_TEAM_WINS"])
```

```python
preds = rf.predict(test[predictors])
```

```python
from sklearn.metrics import accuracy_score
acc = accuracy_score(test["HOME_TEAM_WINS"], preds)
```

```python
acc
```

```python
combined = pd.DataFrame(dict(actual=test["HOME_TEAM_WINS"], prediction=preds), index=test.index)
pd.crosstab(index=combined["actual"], columns=combined["prediction"])
```

```python
from sklearn.metrics import precision_score
```

```python
precision_score(test["HOME_TEAM_WINS"], preds)
```

```python
grouped_games = games.groupby("HOME_TEAM_ID")
group = grouped_games.get_group("GSW")
group
```

```python
def rolling_averages(group, cols, new_cols):
    group = group.sort_values("GAME_DATE_EST")
    rolling_stats = group[cols].rolling(3, closed='left').mean()
    group[new_cols] = rolling_stats
    group = group.dropna(subset=new_cols)
    return group
```

```python
games.dtypes
```

```python
cols = ["PTS_home", "FG3_PCT_home", "FG_PCT_home", "AST_home", "REB_home"]
new_cols = [f"{c}_rolling" for c in cols]
```

```python
new_cols
```

```python
rolling_averages(group, cols, new_cols)
```

```python
games_rolling = games.groupby("HOME_TEAM_ID").apply(lambda x: rolling_averages(x, cols, new_cols))
```

```python
games_rolling
```

```python
games_rolling = games_rolling.droplevel("HOME_TEAM_ID")
```

```python
games_rolling
```

```python
games_rolling.index = range(games_rolling.shape[0])
games_rolling
```

```python
def make_predictions(data, predictors):
    train = data[data["GAME_DATE_EST"] < '2018-01-01']
    test = data[data["GAME_DATE_EST"] >= '2018-01-01']
    rf.fit(train[predictors], train["HOME_TEAM_WINS"])
    preds = rf.predict(test[predictors])
    combined = pd.DataFrame(dict(actual=test["HOME_TEAM_WINS"], prediction=preds), index=test.index)
    precision = precision_score(test["HOME_TEAM_WINS"], preds)
    return combined, precision
```

```python
combined, precision = make_predictions(games_rolling, predictors + new_cols)
precision
```

```python
combined = combined.merge(games_rolling[["GAME_DATE_EST", "HOME_TEAM_ID", "VISITOR_TEAM_ID", "HOME_TEAM_WINS"]], left_index=True, right_index=True)
combined
```

```python
merged = combined.merge(combined, left_on=["GAME_DATE_EST", "VISITOR_TEAM_ID"], right_on=["GAME_DATE_EST", "VISITOR_TEAM_ID"])
```
